# Grafico de Venn

```python
#listagem da coluna 'Nomes' do datafame de alunos
lista_nomes_alunos = list(df_alunos['Nome'])
```

```python
#listagem da coluna 'Nomes' do datafame de onibus
lista_nomes_onibus = list(df_onibus['Nome'])
```

```python
#Importando uma biblioteca grafica (Matplotlib)
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn3
```

```python
#Montando a funcão do diagrama de Venn
def func_venn(venn, lst, labels):

  return venn(lst, set_labels = labels)
```

```python
func_venn(venn2, [set(lista_nomes_alunos), set(lista_nomes_onibus)],
          ("Lista de Nomes dos Alunos", "Lista de Nomes que andam de Onibus"))
plt.title("Diagrama de Venn dos Alunos que andam de Onibus")
plt.show()
```
