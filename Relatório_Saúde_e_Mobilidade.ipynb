```python
#Importando uma biblioteca grafica (Matplotlib)
from matplotlib import pyplot as plt
from matplotlib_venn import venn2, venn3
```

```python
#Montando a funcão do diagrama de Venn
def func_venn(venn, lst, labels):

  return venn(lst, set_labels = labels)
```

```python
func_venn(venn2, [set(lista_nomes_onibus), set(lista_nomes_dengue)],
          ("Lista de Nomes que andam de Onibus", "Lista de Nomes com Dengue"))
plt.title("Diagrama de Venn das pessoas que andaram de Onibus com Dengue")
plt.show()
```
